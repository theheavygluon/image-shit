```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
from PIL import Image
import time
```

```python
##Helper functions

import skvideo.io  
def get_mean(vid):
    return np.mean(skvideo.io.vread(vid),axis=0)

##Convert to BW

def convert_to_bin(img, threshold = 100, cut=0):
    return np.array([1 if i>=int(threshold) else 0 for i in img.reshape(int(img.size),1)]).reshape(1080-cut,1920)

##Flatten

def flatten(img):
    return img[:,  :, 0]

def fast_mean(vid):
    return np.mean(skvideo.io.vread(vid)[ :, :,  :, 0],axis=0)
    
```
